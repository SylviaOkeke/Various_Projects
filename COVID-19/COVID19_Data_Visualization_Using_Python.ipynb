**3.Visualizing how intensive the COVID-19 Transmission has been in Nigeria**

**Task 3.1: Let's see data related to Nigeria**

```python
df_nigeria = df[df.Country == "Nigeria"]
df_nigeria.head()
```

**Task 3.2: Let's select the columns that we need**

```python
df_nigeria = df_nigeria[["Date", "Confirmed"]]
```

```python
df_nigeria.head()
```

**Task 3.3: Let's calculate the first derivation of confirmed column**

```python
df_nigeria["Infection Rate"] = df_nigeria["Confirmed"].diff()
```

```python
df_nigeria.head()
```

**Task 3.4: Let's visualize that relationship**

```python
px.line(df_nigeria, x = "Date", y = ["Confirmed", "Infection Rate"] )
```

**Task 3.5: Let's calculate the maximum infection rate or the maximum number off new infected cases in one day in Nigeria**

```python
df_nigeria["Infection Rate"].max()
```

**Task 3.2: Let's calculate maximum infection rate for all of the countries**

```python
df.head()
```

```python
countries = list(df["Country"].unique())
max_infection_rates = []
for c in countries:
    MIR = df[df.Country == c].Confirmed.diff().max()
    max_infection_rates.append(MIR)
```

**Task 3.3: Let's create a new Dataframe**

```python
df_MIR = pd.DataFrame()
df_MIR["Country"] = countries
df_MIR["Max Infection Rate"] = max_infection_rates
df_MIR.head()
```

**Task 3.4: Let's plot the barchart: maximum infection rate of each country**

```python
px.bar(df_MIR, x="Country", y="Max Infection Rate", color="Country", title="Global Maximum Infection Rate", log_y=True)
```

**4.Let's see how national lockdown impacts COVID-19 transmission in Nigeria**

**COVID-19 pandemic lockdown in Nigeria**

Between February until March 2020, nationwide, homes may only be left for one of the following reasons: necessary professional activities, necessary purchases (groceries or medication), assisting other people, activities outside, alone or in the company of people living in the same household.

**Task 4.1: Lets make variables with the lockdown start date and a month later**

```python
nigeria_lockdown_start_date = '2020-01-22'
nigeria_lockdown_a_month_later = '2020-02-21'
```


**Task 4.2: Let's get data related to Nigeria and check the dataframe**

```python
df_nigeria = df[df.Country == "Nigeria"]
df_nigeria.head()
```

**Task 4.4: Let's do the visualization**

```python
fig = px.line(df_nigeria, x="Date", y="Confirmed", title = "Before and After Lockdown in Nigeria")
fig.add_shape(
    dict(
    type="line",
    x0=nigeria_lockdown_start_date,
    y0=0,
    x1=nigeria_lockdown_start_date,
    y1=df_nigeria["Confirmed"].max(),
    line=dict(color="red", width=2)
    )
)

fig.add_annotation(
    dict(
    x = nigeria_lockdown_start_date,
    y = df_nigeria["Confirmed"].max(),
    text = "Starting date of the lockdown"
    )
)
fig.add_shape(
    dict(
    type="line",
    x0=nigeria_lockdown_a_month_later,
    y0=0,
    x1=nigeria_lockdown_a_month_later,
    y1=df_nigeria["Confirmed"].max(),
    line=dict(color="black", width=2)
    )
)

fig.add_annotation(
    dict(
    x = nigeria_lockdown_a_month_later,
    y = df_nigeria["Confirmed"].max(),
    text = "A month later"
    )
)
```

**Task 5.5: Let's normalize the columns**

```python
df_nigeria["Infection Rate"] = df_nigeria["Infection Rate"]/df_nigeria["Infection Rate"].max()
df_nigeria["Deaths Rate"] = df_nigeria["Deaths Rate"]/df_nigeria["Deaths Rate"].max()
```

**Task 5.6: Let's plot the line chart again**

```python
fig = px.line(df_nigeria, x="Date", y=["Infection Rate", "Deaths Rate"])

fig.add_shape(
    dict(
    type="line",
    x0=nigeria_lockdown_start_date,
    y0=0,
    x1=nigeria_lockdown_start_date,
    y1=df_nigeria["Infection Rate"].max(),
    line=dict(color="black", width=3)
    )
)

fig.add_annotation(
    dict(
    x = nigeria_lockdown_start_date,
    y = df_nigeria["Infection Rate"].max(),
    text = "Starting date of the lockdown"
    )
    )

fig.add_shape(
    dict(
    type="line",
    x0=nigeria_lockdown_a_month_later,
    y0=0,
    x1=nigeria_lockdown_a_month_later,
    y1=df_nigeria["Infection Rate"].max(),
    line=dict(color="black", width=3)
    )
    )

fig.add_annotation(
    dict(
    x = nigeria_lockdown_a_month_later,
    y = df_nigeria["Infection Rate"].max(),
    text = "A month later"
    )
)
fig.show()
```
